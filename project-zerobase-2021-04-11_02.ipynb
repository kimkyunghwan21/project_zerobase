```python
from urllib.request import Request, urlopen
from urllib.parse import urlencode, quote_plus
import xml.etree.ElementTree as ET
import xmltodict
import json
from bs4 import BeautifulSoup
import pandas as pd

 

url = 'http://apis.data.go.kr/B551182/hospInfoService/getHospBasisList?ServiceKey=qN8YHL9uF%2FU9hr%2BHUmV5PRas7zAcxkH%2F06yQzqieyJ%2Fvd21j3QfenpwjNALvLP71D1pjTIhQAUdY%2FJGHguDiVg%3D%3D&pageNo={}'

 

for page in range(1, 9999999) :

    request = Request(url.format(page))
    request.get_method = lambda: 'GET'
    response_body = urlopen(request).read()

    if(request.status_code == 404 ):
        print(page)
        break
 
    string_xml = response_body.decode('utf-8')
    #print(string_xml)
    tree = ET.fromstring(string_xml)
    #print(tree)
    ET.dump(tree)   #메모리에 xml 객체 트리 구조 생성

 
    print()    

    df = pd.DataFrame(columns=['xpos', 'ypos','yadmnm', 'addr', 'clcdnm', 'hospurl','drtotcnt', 'resdntcnt' ,  'intncnt',  'gdrcnt', 'sdrcnt'])

    i=0

 

    soup = BeautifulSoup(string_xml, "html.parser")
    items = soup.find("items")
    #print(type(items))
    #print(items)

   

    for item in items.findAll("item"):    

        df.loc[i, 'xpos'] =  item.xpos.text
        df.loc[i, 'ypos'] =  item.ypos.text
        df.loc[i, 'yadmnm'] = item.yadmnm.text
        df.loc[i, 'addr'] =  item.addr.text
        df.loc[i, 'clcdnm'] = item.clcdnm.text

        if item.hospurl == None:
            df.loc[i, 'hospurl'] = ' '

        else :

            df.loc[i, 'hospurl'] = item.hospurl.text 

        df.loc[i, 'drtotcnt'] = item.drtotcnt.text
        df.loc[i, 'resdntcnt'] = item.resdntcnt.text
        df.loc[i, 'intncnt'] =item.intncnt.text
        df.loc[i, 'gdrcnt'] = item.gdrcnt.text
        df.loc[i, 'sdrcnt'] = item.sdrcnt.text

        i=i+1

        print(item)

 

    df.to_csv('hostpitals02.csv', mode = 'a', encoding='euc-kr', index=False)

     
```
